# Data preposesing

## Encoding the categorical features

```python
# encoding the sex column

from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
lc.fit(df['sex'])
df['sex']=lc.transform(df['sex'])

```

```python
# encoding the region column

from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
lc.fit(df['region'])
df['region']=lc.transform(df['region'])
```

```python
# encoding smoker column

from sklearn.preprocessing import LabelEncoder
lc=LabelEncoder()
lc.fit(df['smoker'])
df['smoker']=lc.transform(df['smoker'])
```

```python
df.head()
```

```python

import matplotlib.pyplot as plt
# checking for outliers
plt.figure(figsize=(20,15))
df.boxplot()
```

```python


Q1 = df.charges.quantile(0.25)
Q3 = df.charges.quantile(0.75)
IQR = Q3 - Q1
df = df[~((df.charges < (Q1 - 1.5 * IQR)) | (df.charges > (Q3 + 1.5 * IQR)))]
```

```python


import matplotlib.pyplot as plt
plt.boxplot(df['charges'])
plt.title('charges boxplot')
plt.show()
```

## splitting target and features

```python
x = df.drop('charges',axis=1)
y = df['charges']
```

```python
x
```

```python
y
```

# traing and testing

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# standization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

###HYPER PARAMETER TUNING

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [2, 4, 6, 8, 10],
    'min_samples_split': [2, 5, 10, 15, 20],
    'min_samples_leaf': [1, 2, 4, 6, 8, 10]
}
```

```python
dt1=DecisionTreeRegressor(max_depth=4)
```

```python
grid_search = GridSearchCV(estimator=dt1, param_grid=param_grid, cv=5)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```

```python
best_dt = grid_search.best_estimator_
y_pred = best_dt.predict(x_test)
print("Test R2 score: ", metrics.r2_score(y_test, y_pred))
```

here we can says that the model is hyper parameter tunned for the best combination of parameter and give better accuracy that default parameter
